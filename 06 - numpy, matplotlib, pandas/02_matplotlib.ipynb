## matplotlib
Matplotlib is a python library used to create 2D graphs and plots by using python scripts. It has a module named pyplot which makes things easy for plotting by providing feature to control line styles, font properties, formatting axes etc. It is used along with NumPy to provide an environment that is an effective open source alternative for MatLab.

```python
from matplotlib import pyplot as plt

%matplotlib inline

y = [2, 4, 6, 4]
plt.plot(y)           # only one list provided, so it will take them as y values and
                      # x values will be corresponding indices.
plt.ylabel('numbers')
plt.xlabel('indices')
plt.title('my plot')

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y, linewidth=4)    # linewidth sets line's width
plt.xlabel('numbers')
plt.ylabel('squares')
plt.title('Squares')
plt.grid()      # grids On

plt.show()
```

```python
# plotting points
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y, 'ro')
plt.grid()
plt.show()
```

```python
import numpy as np

n = np.arange(0, 3, 0.2)

# blue dashes, red squares, green triangles, yellow dots
plt.plot(n, n, 'b--', label='^1')
plt.plot(n, n**2, 'rs', label='^2')
plt.plot(n, n**2.5, 'g^', label='^2.5')
plt.plot(n, n**3, 'yo', label='^3')
plt.grid()       # grid On
plt.legend()     # legends On

plt.show()
```

```python
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
x2 = [1, 2, 3, 4]
y2 = [1, 8, 81, 256]

# plot() returns lists of lines
lines = plt.plot(x1, y1, x2, y2)

# to set parameters
plt.setp(lines[0], color='r')
plt.setp(lines[1], color='b')

plt.grid()
plt.show()
```

```python
n = np.arange(0, 20, 0.1)

# selects figure - 1
plt.figure(1)
# subplot() takes 3 digit number as parameter to specify which plot among the subplot
# each digit corresponds to row, column, sub-figure respectively
plt.subplot(211)
plt.plot(n, np.sin(n), 'b')
plt.grid()
plt.subplot(212)
plt.plot(n, np.cos(n), 'r')
plt.grid()

plt.show()
```

```python
# select figure - 1
plt.figure(1)
# select subplot
plt.subplot(211)
plt.plot([1, 2, 3, 4])
# select subplot
plt.subplot(212)
plt.plot([4, 3, 2, 1])

# select figure - 2
plt.figure(2)
plt.plot([3, 4, 5, 6])

# select figure - 1
plt.figure(1)
# select subplot
plt.subplot(211)
plt.title('testing')

plt.show()
```
